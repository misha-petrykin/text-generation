```python
from xml.dom.minidom import parseString
from bs4 import BeautifulSoup
import pandas as pd
import dicttoxml
import requests
import re
```

```python
url='https://nure.ua/ru/'

r=requests.get(url).text
soup=BeautifulSoup(r,'lxml')

l=soup.find_all('div',class_='sub-article-cover col-md-3 col-sm-6 col-xs-12 anim animate-fade animate-bottom')
```

```python
for i in l:
    print(i.text+'\n--------------\n')
```

```python
def news(i):
    d=dict()
    d['Заголовок']=i.find('a',class_='title').text
    d['Дата']=i.find('span',class_='date').text
    d['Текст']=i.find('p').text
    return d
s=dict()
for i,j in enumerate(l):
    s['Новости id="'+str(i+1)+'"']=news(j)
print(s['Новости id="1"'])
s
```
